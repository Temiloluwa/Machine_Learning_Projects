# Wine Quality Dataset Solution

## 1. Data Preprocessing

### 1.1 Reading the Dataset

- red wine quality data is read and stored in variable `winred`
- white wine quality data is read and stored in variable `winwhite`

```python
winred  = pd.read_csv("winered.csv", delimiter = ";")
winwhite  = pd.read_csv("winewhite.csv", delimiter = ";")
winred.head()
```

```python
winwhite.head()
```

- create new column `winetype` in the two datasets
- concatenate the two datasets
- onehot encode column `winetype`
- red are white are stored jointly as `wine_data`

```python
#create new colum winetype in the two datasets
winred["winetype"] = 'red'
winwhite["winetype"] = 'white'

wine_data = winred

#concatenate the two datasets
wine_data = pd.concat([winred,winwhite])
wine_data = pd.get_dummies(wine_data)

wine_data.head()
```

```python
wine_data.describe()
```
